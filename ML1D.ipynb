```python
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D


import numpy as np
import scipy as sc
from scipy.stats import multivariate_normal
plt.rcParams.update({'font.size': 20})
plt.rcParams['figure.figsize'] = [12, 10]
import matplotlib.image as mpimg
```

## Definition of the 1-D Gaussian function


```python
def Gaussian(x,mu,sigma):
    Delta_squared = np.square((x-mu)/sigma)
    P = 1/np.sqrt(2*np.pi*np.square(sigma))*np.exp(-Delta_squared/2)
    return P
```

## Example 1: Estimate the mean of a Gaussian random variable

```python
mu_orig = 1
sigma = 0.3
Nsamples = 50

Ngrid = 100
mugrid = np.linspace(-4, 4, Ngrid)

x = mu_orig + np.random.normal(0,sigma,Nsamples)
likelihood = np.ones(np.shape(mugrid))
#plt.plot(grid,Gaussian(grid,mu,sigma))

for i in range(Nsamples):
    g = Gaussian(x[i],mugrid,sigma)
    plt.plot(mugrid, g,'k:',linewidth=0.5)
    likelihood = likelihood*g
    
likelihood = 1.4*likelihood/max(likelihood)
s = plt.plot(mugrid, likelihood,'r',linewidth=3,label='Joint Likelihood')

s = plt.title('Nsamples='+str(Nsamples))
s = plt.xlabel("$\mu$")

# CHANGE 0 to YOUR CODE
#--------------------------

mlsolution = np.mean(x)
#--------------------------
s = plt.plot(mlsolution,1.4,'mo',markersize=30,label='ML solution')
s = plt.legend()
```

## Log likelihood plots

```python
mu_orig = 1
sigma = 0.3
Nsamples = 100

Ngrid = 100
mugrid = np.linspace(-4, 4, Ngrid)

x = mu_orig + np.random.normal(0,sigma,Nsamples)
likelihood = np.ones(np.shape(mugrid))
#plt.plot(grid,Gaussian(grid,mu,sigma))

for i in range(Nsamples):
    g = Gaussian(x[i],mugrid,sigma)
    plt.plot(mugrid, np.log10(g),'k:',linewidth=0.5)
    likelihood = likelihood*g
    
likelihood = 1.4*likelihood/max(likelihood)
s = plt.plot(mugrid, np.log10(likelihood),'r',linewidth=3,label='Joint Likelihood')
plt.title('Nsamples='+str(Nsamples))
plt.xlabel("$\mu$")

# CHANGE 0 to YOUR CODE
#--------------------------

mlsolutionnew = np.mean(x)
#--------------------------
s = plt.plot(mlsolutionnew,0,'mo',markersize=30,label='ML solution')
s = plt.legend()

#plt.savefig(string)
```
